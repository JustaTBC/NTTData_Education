```sql
USE CATALOG nttdataeducacao;
USE SCHEMA gold;

SELECT
    -- 1. Chave de agrupamento
    area_extensa,

    -- 2. Totais gerais
    total_itens,
    total_anulados,
    total_invalidos,
    total_itens_adaptados,

    -- 3. Médias dos parâmetros TRI
    media_dificuldade,
    media_discriminacao,
    media_acerto_ao_acaso,

    -- 4. Contagem por classificação de dificuldade
    qtd_muito_facil,
    qtd_facil,
    qtd_medio,
    qtd_dificil,

    -- 5. Proporções gerais
    proporcao_anulados,
    proporcao_adaptados

FROM nttdataeducacao.gold.agg_itens_por_area
ORDER BY total_itens DESC;
```
