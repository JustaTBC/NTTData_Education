```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, to_date, year, lit,
    when, trim, regexp_replace
)
from pyspark.sql.types import (
    IntegerType, DoubleType, StringType
)
```

```python
df_silver = spark.sql("select * from nttdataeducacao.silver.escolas_2024")
display(df_silver)
```

```python
from pyspark.sql.functions import col, sum, avg, countDistinct, count
from pyspark.sql.types import DoubleType

print("Criando tabela Gold: CADASTRO_ESCOLAS")
df_cadastro_escolas = df_silver.select(
    col("cod_entid").alias("codigo_escola"), 
    col("nome_entid").alias("nome_escola"),
    col("cod_munic").alias("codigo_municipio"),
    col("nome_munic").alias("nome_municipio"),
    col("uf"),
    col("regiao"),
    col("localizacao"),
    col("dependencia")
).dropDuplicates(["codigo_escola"]) 
```

```python
df_distribuicao = df_silver.groupBy(
    col("uf"),
    col("localizacao"),
    col("dependencia")
).agg(
    count(col("cod_entid")).alias("codigo_da_escola")
).orderBy(
    col("uf"),
    col("codigo_da_escola").desc() # Ordena para ver as maiores contagens primeiro
)
```

```python
display(df_distribuicao)
```

```python
df_distribuicao.write.mode("overwrite").saveAsTable("nttdataeducacao.gold.qnt_escolas_localizacao")
```
