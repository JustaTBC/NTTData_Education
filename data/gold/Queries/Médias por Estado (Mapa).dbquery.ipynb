```sql
SELECT 
    -- 1. Chave do Mapa
    exam_state,

    -- 2. Contexto (Quantos alunos)
    total_inscritos,

    -- 3. A Média Geral (Dividida por 5, pois veio da soma 0-5000)
    (media_geral / 5) as media_geral_enem,

    -- 4. As Matérias Específicas (NÃO DIVIDE, já são 0-1000)
    media_redacao,
    media_matematica,
    media_linguagens,
    media_humanas,
    media_natureza

FROM nttdataeducacao.gold.agg_performance_estado

ORDER BY media_geral_enem DESC
```
