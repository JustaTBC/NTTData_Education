```sql
SELECT 
    -- 1. O Nome Real da Escola (Fica perfeito no eixo Y)
    nome_escola as escola_label,
    
    -- 2. A Cidade (Trouxe separada caso queira mostrar no mouse-over)
    school_city,

    -- 3. A nota média na escala 0-1000
    -- Pegamos o Total (que é ~4000) e dividimos por 5
    round(Nota_Total_Ranking / 5, 2) as media_padrao_enem

FROM nttdataeducacao.gold.agg_ranking_top10_final

ORDER BY media_padrao_enem DESC
```
