```sql
SELECT 
    -- 1. IMPORTANTE: Traga o estado para o filtro funcionar (pode deixar escondido)
    school_state, 

    -- 2. O Label (Nome da Escola)
    concat(school_city, ' (ID: ', cast(school_id as string), ')') as escola_label,

    -- 3. As Matérias (Empilhadas)
    round(media_escola_redacao, 0) as Redacao,
    round(media_escola_matematica, 0) as Matematica,
    round(media_escola_linguagens, 0) as Linguagens,
    round(media_escola_humanas, 0) as Humanas,
    round(media_escola_natureza, 0) as Natureza,
    
    -- 4. O Total (Para ordenar)
    media_escola_total

FROM nttdataeducacao.gold.agg_ranking_escolas
WHERE school_id IS NOT NULL

-- O Filtro do Dashboard vai ser injetado automagicamente AQUI pelo Databricks

ORDER BY media_escola_total DESC
LIMIT 10
```
