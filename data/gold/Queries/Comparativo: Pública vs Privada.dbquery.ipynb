```sql
SELECT 
    -- 1. Eixo X: O Tipo de Escola (Já vem pronto dessa tabela)
    tipo_escola,
    
    -- 2. Contexto: Quantidade de Alunos (Para Tooltip)
    qtd_alunos,

    -- 3. Eixo Y: Médias Ajustadas (Dividido por 5 para escala 0-1000)
    (media_geral / 5) as media_geral_enem,
  media_redacao, 
  media_matematica,
  media_escola_linguagens as media_linguagens, -- Renomeando pra ficar curto
    media_escola_humanas as media_humanas,
    media_escola_natureza as media_natureza

FROM nttdataeducacao.gold.agg_publica_vs_privada

-- Opcional: Filtra "Outros" se quiser focar só nas principais
WHERE tipo_escola <> 'Outros'

ORDER BY media_geral_enem DESC
```
