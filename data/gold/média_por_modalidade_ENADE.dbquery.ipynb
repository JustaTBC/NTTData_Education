```sql
--Media por modalidade
SELECT
  modalidade_de_ensino,
  ROUND(AVG(media_enade_continuo),3) AS media_modalidade,
  COUNT(DISTINCT codigo_do_curso) AS qtd_cursos
FROM nttdataeducacao.gold.enade_curso
WHERE modalidade_de_ensino IS NOT NULL
  AND (:uf = 'ALL' OR sigla_da_uf = :uf)
  AND (:area = 'ALL' OR area_de_avaliacao = :area)
GROUP BY modalidade_de_ensino
ORDER BY media_modalidade DESC;
```
