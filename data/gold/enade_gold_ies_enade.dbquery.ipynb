```sql
SELECT
  concat(nome_da_ies, ' — ', codigo_do_curso) AS curso_label,
  ROUND(media_fg,3) AS nota_fg,
  ROUND(media_ce,3) AS nota_ce
FROM nttdataeducacao.gold.enade_curso
WHERE (:uf = 'ALL' OR sigla_da_uf = :uf)
  AND (:area = 'ALL' OR area_de_avaliacao = :area)
ORDER BY nota_fg DESC
LIMIT 20;
SELECT 
  COUNT(DISTINCT codigo_do_curso) AS qtd_cursos
FROM nttdataeducacao.gold.enade_curso
WHERE (:uf = 'ALL' OR sigla_da_uf = :uf)
  AND (:area = 'ALL' OR area_de_avaliacao = :area)
  AND (:modalidade = 'ALL' OR modalidade_de_ensino = :modalidade);

-- Nº total de alunos valor unico
SELECT 
  COALESCE(
    SUM(qtd_alunos),
    NULL
  ) AS total_alunos
FROM nttdataeducacao.gold.enade_curso
WHERE (:uf = 'ALL' OR sigla_da_uf = :uf)
  AND (:area = 'ALL' OR area_de_avaliacao = :area)
  AND (:modalidade = 'ALL' OR modalidade_de_ensino = :modalidade);

--coluna calculada
SELECT
  nome_da_ies || ' - ' || codigo_do_curso AS curso_label,
  media_enade_continuo
FROM nttdataeducacao.gold.enade_curso
ORDER BY media_enade_continuo DESC
LIMIT 20;

--Top N Cursos por Média ENADE, grafico de barras 
SELECT
  nome_da_ies,
  codigo_do_curso,
  area_de_avaliacao,
  modalidade_de_ensino,
  ROUND(media_enade_continuo,3) AS media_enade
FROM nttdataeducacao.gold.enade_curso
WHERE (:uf = 'ALL' OR sigla_da_uf = :uf)
  AND (:area = 'ALL' OR area_de_avaliacao = :area)
  AND (:modalidade = 'ALL' OR modalidade_de_ensino = :modalidade)
ORDER BY media_enade DESC
LIMIT CAST(:top_n AS INT);

--Comparação FG x CE
SELECT
  concat(nome_da_ies, ' — ', codigo_do_curso) AS curso_label,
  ROUND(media_fg,3) AS nota_fg,
  ROUND(media_ce,3) AS nota_ce
FROM nttdataeducacao.gold.enade_curso
WHERE (:uf = 'ALL' OR sigla_da_uf = :uf)
  AND (:area = 'ALL' OR area_de_avaliacao = :area)
ORDER BY nota_fg DESC
LIMIT 20;
```
